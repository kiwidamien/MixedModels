## Background

We are going to look at the test scores of different dragons, where the dragons come from similar mountain ranges. We notice that there are repeated measurements on each site.

The website that has the tutorial (in R) is [here](https://gkhajduk.github.io/2017-03-09-mixed-models/)

While it models a different problem, [this example on doctors](https://stats.idre.ucla.edu/other/mult-pkg/introduction-to-linear-mixed-models/) has a nice formal description of mixed effects models.

````python
dragons.testScore.hist(bins=20)
````

Let's make a simple linear regression.

````python
simple_model = smf.ols(
    formula='testScore ~ bodyLength',
    data = dragons
)

result = simple_model.fit()

result.summary()
````

````python
predictions = result.predict(exog=dragons)

plt.scatter(dragons.bodyLength, dragons.testScore, c='r', label='observed')
plt.plot(dragons.bodyLength, predictions, c='b', label='predictions')
````

````python
(ggplot(dragons, aes(x = 'bodyLength', y = 'testScore')) +
  geom_point() +
  geom_smooth(method = "lm") )
````

Let's look at what happens if we look at each individual mountain range by using the hue:

````python
(ggplot(dragons, aes(x = 'bodyLength', y = 'testScore', colour='mountainRange')) +
  geom_point() +
  geom_smooth(method = "lm") )
````

We see that the variations in the best fit lines are huge due to the mountain ranges. Instead we could look at what making an intercept for each mountain range would do if we added it as a categorical variable:

````python
mountain_range_model = smf.ols(formula='testScore ~ bodyLength + mountainRange',
                               data=dragons)

results_mountain = mountain_range_model.fit()

print(results_mountain.summary())
````

We see that the coefficient of `bodyLength` is not significant, as it straddles zero.

The blog post argues if our question of interest is "Are `bodyLength` significant?" then we want to control for the variation due to different mountain ranges; we are not specifically interested in determining what the coefficents are. So we would be better off looking at mixed effects models.

## Our first mixed effect model

We are going to try and estimate a fixed effect (bodyLength) against a random effect (mountainRange). Specifically, we are going to allow each mountain range to have it's own intercept.

The patsy model (from R / `lme4`) we are trying to model is

```python
testScore ~ bodyLength + (1|mountainRange)
```


````python
# Build the model
mlm_mod = sm.MixedLM.from_formula(
    formula = 'testScore ~ bodyLength', 
    groups = 'mountainRange', 
    re_formula="1", 
    data=dragons
)

# Run the fit
mlm_result = mlm_mod.fit()

# Print out the summary of the fit
mlm_result.summary()
````

````python
339.666/dragons.testScore.var()
````

We do see a difference in the confidence intervals in the `bodyLength` is different in each model. The models are not guaranteed to get the same results.

For everything as a fixed effect, we have the slope for the `bodyLength` as $(-0.144, 0.169)$

For the random effect, we have the slope for the `bodyLength` as $(-0.122, 0.189)$

### Model specification

We have a general model specification as
$$
y = X\beta + Z u + \epsilon
$$

where 
* $X$ is $N \times p$, where $p$ is the number of fixed effects (these are our fixed variables)
* $\beta$ is $p \times 1$, the coefficients of fixed parameters
* $Z$ is $N \times qJ$, where $J$ is the number of groups and $q$ is the number of features per group
* $u$ are the random variables $qJ \times 1$
* $\epsilon$ is the irreducible error

The [post](https://stats.idre.ucla.edu/other/mult-pkg/introduction-to-linear-mixed-models/) does a really nice job walking through this.

What does this look like for our model? In our case, we have $N=480$ (the total number of observations we have). We have $J=8$ mountain ranges, and for each one we are estimating one parameter ($q=1$) -- the intercept. The matrix $X$ will have two columns: a column of 1's (for the intercept), and a column with the fixed effect of `bodyLength`

````python
def ols_model(d):
    lengthSlope = numpyro.sample('lengthSlope', dist.Normal(0,15))
    intercept = numpyro.sample('intercept', dist.Normal(0, 90))
    stdError = numpyro.sample('epsilon', dist.Normal(0,30))
    
    pt_estimate = intercept*1 + lengthSlope*d.bodyLength.values
    numpyro.sample('obs', dist.Normal(pt_estimate, stdError), obs=d.testScore.values)
````

````python
rng_key = random.PRNGKey(0)
rng_key, rng_key_ = random.split(rng_key)

# Run NUTS.
kernel = NUTS(ols_model)
num_samples = 2000
mcmc = MCMC(kernel, num_warmup=1000, num_samples=num_samples)
mcmc.run(rng_key_, 
         d=dragons,
        )

samples_1 = mcmc.get_samples()
````

````python
mcmc.print_summary()
````

This matches pretty well to what we got when we ignored the different site using OLS. Now let's try making the mixed model specification

````python
def mixed_model(d):
    lengthSlope = numpyro.sample('lengthSlope', dist.Normal(0,15))
    intercept = numpyro.sample('intercept', dist.Normal(0, 90))
    stdError = numpyro.sample('epsilon', dist.Normal(0,30))
    
    n_ranges = d.mountainRange.nunique()
    mountain_range_index = d.mountainRange.astype('category')
    
    u_range = numpyro.sample('u_range', dist.Normal(0, 60), sample_shape=(n_ranges,))
    
    pt_estimate = (
        (intercept*1 + lengthSlope*d.bodyLength.values) +  # X beta
        (u_range[mountain_range_index.cat.codes.values])   # Z u
    )
    numpyro.sample('obs', dist.Normal(pt_estimate, stdError), obs=d.testScore.values)
````

As described in the blog post above, if we just have a random intercept per mountainRange, then Z has 8 columns (one for each mountain range). 

$Z_{i,j}$ is 1 if dragon $i$ is in range $j$, and zero otherwise. 

$u_j$ is the (random) value of the intercept for range $j$

So $Z u$ selects which of the 8 values of $u$ we use (we use the value corresponding to that dragon's mountain range). Our way of writing it makes this explicit.

````python
rng_key = random.PRNGKey(0)
rng_key, rng_key_ = random.split(rng_key)

# Run NUTS.
kernel = NUTS(mixed_model)
num_samples = 2000
mcmc = MCMC(kernel, num_warmup=1000, num_samples=num_samples)
mcmc.run(rng_key_, 
         d=dragons,
        )

samples_1 = mcmc.get_samples()
````

````python
mcmc.print_summary()
````

````python
# Let's take the variance of the U component two different ways
# 1. Variance of the means
# 2. Variance of the individual points
````

````python
samples_1['u_range'].var()
````

The mountain range variance from the mixed effects model was approximately 339, which is between these values. The error between a variance of 300 and 330 isn't too bad, especially when converted to standard deviation.
